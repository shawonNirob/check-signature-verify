```python
import pytesseract
from PIL import Image, ImageFilter, ImageOps
import re
import json

# specify tesseract path
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\shawon.sarowar\AppData\Local\Programs\Tesseract-OCR\tesseract.exe'

# Load image using PIL
img_path = "C:/Git/LLM/llama/bank_cheque.JPG"
image = Image.open(img_path)

# Preprocess image using PIL
gray = ImageOps.grayscale(image) #Converts the image to grayscale to reduce noise and improve OCR accuracy.
gray = gray.filter(ImageFilter.MedianFilter(size=3))  # Reduces noise by smoothing the image (similar to a bilateral filter in OpenCV).

# OCR text extraction
text = pytesseract.image_to_string(gray)

# Print the full OCR text (for debugging)
print("\n--- OCR Output ---\n")
print(text)

# Regex extraction rules
def extract_bank_name(text):
    match = re.search(r'\b([A-Za-z&]+\s+Bank(?:\s+(?:of\s+\w+|PLC|Ltd|Limited))?)\b', text, re.IGNORECASE)
    return match.group(1).strip() if match else None

def extract_account_number(text):
    match = re.search(r'\b\d{10,16}\b', text)
    return match.group(0) if match else None

def extract_date(text):
    match = re.search(r'\b(\d{2}[-/]\d{2}[-/]\d{4}|\d{8})\b', text)
    return match.group(1) if match else None


# Apply extraction
extracted_data = {
    "bank_name": extract_bank_name(text),
    "account_number": extract_account_number(text),
    "date": extract_date(text)
}

print("\n--- Extracted Info ---\n")
#print(extracted_data)
print(json.dumps(extracted_data, indent=2))
```

```python
import pytesseract
from PIL import Image
import cv2
import re
import json

# If on Windows, specify tesseract path
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\shawon.sarowar\AppData\Local\Programs\Tesseract-OCR\tesseract.exe'

# Load image
img_path = "C:/Git/LLM/llama/bank_cheque.JPG"
image = cv2.imread(img_path)

# Optional: preprocess for better OCR accuracy
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.bilateralFilter(gray, 11, 17, 17)

# OCR text extraction
text = pytesseract.image_to_string(gray)

# Print the full OCR text (for debugging)
print("\n--- OCR Output ---\n")
print(text)

# Regex extraction rules
def extract_bank_name(text):
    # Assume bank name is in uppercase and includes "BANK"
    match = re.search(r'([A-Z ]+BANK[A-Z ]*)', text)
    return match.group(1).strip() if match else None

def extract_account_number(text):
    match = re.search(r'Acc(?:ount)?(?: No)?[:\s]*([A-Za-z]?\d{6,})', text, re.IGNORECASE)
    return match.group(1).strip() if match else None

def extract_date(text):
    match = re.search(r'\b(\d{2}[-/]\d{2}[-/]\d{4})\b', text)
    return match.group(1) if match else None

# Apply extraction
extracted_data = {
    "bank_name": extract_bank_name(text),
    "account_number": extract_account_number(text),
    "date": extract_date(text)
}

print("\n--- Extracted Info ---\n")
print(json.dumps(extracted_data, indent=2))
```
